**KMEANS**

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
```

```python
X,_=make_moons(n_samples=200)
```

```python
km.fit(X)
```

```python
print('Data points after clustered are',km.labels_)
```

```python
import pandas as pd
pd.DataFrame(km.labels_).value_counts()
```

```python
print('cluster centroids are',km.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=km.labels_,cmap='viridis')
```

**DBSCAN**

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,_=make_moons(n_samples=200)
```

```python
dbs=DBSCAN(min_samples=5,eps=0.1)
```

```python
dbs.fit(X)
```

```python
dbs.labels_
```

```python
import pandas as pd
pd.DataFrame(km.labels_).value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=dbs.labels_,cmap='viridis')
```

**Agglomerative clustering**

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering()
```

```python
from sklearn.datasets import make_moons
```

```python
X,_=make_moons(n_samples=10)
```

```python
ac.fit(X)
```

```python
ac.labels_
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=ac.labels_,cmap='viridis')
```

```python
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
distances=pdist(X)
```

```python
z=linkage(distances,method='single')
```

```python
dendrogram(z)
```
